```python
#how many climbers in each main area?
f,ax=plt.subplots()
plotdf=cldf[cldf['region']=='California'].groupby('mainarea').count()
plotdf.name.plot(kind='bar', ax=ax, grid=False)
labels=[plotdf.index.values][0]
x=ax.set_xticklabels([adf[adf['areaid']==float(a)].name.values[0] for a in labels])
ax.set_ylabel('# climbers based in area')
sns.despine()
```

### Characterize differe climbing areas

The structure of mountain project is based on a nested hierarchy of areas (e.g. a state is an area, a park is a subarea within the state, individual mountains are subareas within parks, and individual walls are subareas within mountains. However, the user generally thinks about climbing destinations at the level of "park" or large scale "main" area. I therefore focus many of my analyses on "mainareas", operationalized as any area which is a direct descendant of a US state. I plot the number of climbs and areas by region to get a sense of how climbs are distributed across the country. 

```python
#how many subareas are within each area
f,ax=plt.subplots(figsize=[12,4])
counts=adf[adf['region']!='World'].groupby('mainarea').count()
counts.name.plot(kind='bar', ax=ax, grid=False)
labels=[adf.loc[int(v),'name'] for v in counts.index.values if v in adf.index.values]
x=ax.set_xticklabels(labels)
sns.despine()
ax.set_ylabel('number of subareas')
```

```python
#how many individual routes/problems are in each area
f,ax=plt.subplots(figsize=[12,4])
cdf.groupby('mainarea').count().name.plot(kind='bar', ax=ax, grid=False)
labels=[adf.loc[int(v),'name'] if v in adf.index.values else '' for v in cdf.groupby('mainarea').count().index.values]
x=ax.set_xticklabels(labels)
ax.set_ylabel('number of climbs in area')
sns.despine()
```

```python
#total number of climbers who haved climbed in the park
f,ax=plt.subplots(figsize=[12,4])
climbcounts=hdf.groupby('climb').count()
climbcounts['climb']=climbcounts.index.values
plotdf=pd.merge(climbcounts, cdf, left_on='climb', right_on='climbid', how='left')[['mainarea','region', 'climb','hitsid','climber','area']]
plotdf.groupby('mainarea').sum().climber.plot(kind='bar', ax=ax, grid=False)
labels=[adf.loc[int(v),'name'] for v in plotdf.groupby('mainarea').sum().index.values]
x=ax.set_xticklabels(labels)
ax.set_ylabel('# climbers who have climbed in park')
sns.despine()
```

### Relationships between areas

Do certain climbers tend to climb at certain areas? Is there much cross-area climbing (e.g. users climbing at multiple climbing areas)? Some cross-area climbing will be necessary to build personalized recommendations at nonlocal areas using user data. If cross-area climbing is insufficient we may need to rely on metadata and climb attributes to provide personalized recommendations for new areas. To assess this, we identified a "home crag" for each climber, defined as the area in which they interact with the largest number of climbs. Then, for each area we can determine how many climbers interact with the area in total, and what proportion of those are "local" climbers (for whom it is the main crag), and what proportion are "foreign" climbers who have a different dominant area. I also computed the similarity of areas in terms of their climber interactions. Unsurprisingly, this similarity structure was largely determined by geography.

### Do ratings for an individual climb vary across individuals? (i.e. do we need personalization)

```python
f,ax=plt.subplots()
means=sdf.groupby('climb')['starsscore'].mean()
stds=sdf.groupby('climb')['starsscore'].std()
stds.hist(bins=20, grid=False)
sns.despine()
ax.set_ylabel('frequency')
ax.set_xlabel('standard deviation of star rating')
```

```python
#do people differ in their average ratings?
useravg=sdf[sdf['starsscore']>0]
useravg=useravg.groupby('climber').mean()[['starsscore']]
useravg['climber']=useravg.index.values
useravg=useravg.rename(columns={'starsscore':'useravg'})
plotdf=pd.merge(sdf, useravg, on='climber', how='right')
plotdf['useravg'].hist(grid=False)
sns.despine()
plt.ylabel('frequency')
plt.xlabel('average rating by individual user')
```
