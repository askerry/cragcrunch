```python
def getregion(areaid, adf=None):
    areaid=float(areaid)
    if ~np.isnan(areaid) and areaid in adf.areaid.values:
        x=adf[adf['areaid']==float(areaid)]
        return x.region.values[0]
    else:
        return np.nan
udf['region']=udf['mainarea'].apply(getregion, adf=adf)
```

```python
import scipy.stats
def getclimberareas(hitsdf, climbdf, climberdf):
    for c in hitsdf.climber.unique():
        climbids=hitsdf.loc[hitsdf['climber']==c,'climb'].values
        climbids=[c for c in climbids if c in climbdf.climbid.values]
        areas=climbdf.loc[climbids,'mainarea'].values
        numareas=len(set(areas))
        try:
            m=scipy.stats.mode(areas)[0][0]
        except:
            m=np.nan
        climberdf.loc[c,'mainarea']=m
        climberdf.loc[c,'numareas']=numareas
    return climberdf
```

```python
def visualizearea(df):
    f,ax=plt.subplots(1,2, figsize=[8,3])
    df.groupby('style').count()['climbid'].plot(kind='pie', ax=ax[0])
    ax[0].set_ylabel("")
    ax[0].set_title('climb type breakdown')
    df.groupby('numgrade').count()['climbid'].plot(kind='bar', ax=ax[1])
    ax[1].set_ylabel("")
    ax[1].set_title('climb grade breakdown')
    plt.show()
```
