###Validation approach

I will use a variety of algorithms and feature sets to predict individual user ratings. To validate the model, I use leave-one-out cross validation. Can I predict the rating that an individual climber gives a climb based on their ratings of other climbs? Leave-on-out cross-validation has known [limitations](http://www.russpoldrack.org/2012/12/the-perils-of-leave-one-out.html) but is useful in the present context where there can be a very small number of exemplars for each individual climber. On each iteration, I train on all climbs but one, and predict the user's remaining climb.

###Feature selection

To be able to quickly incorporate new users who don't have a full climbing history, I want to identify a relatively simple, intuitive feature vector that can be used to provide predictions for a new user based on a few indicated preferences. So, I want to reduce this space of features. Many approaches to dimensionality reduction (i.e. PCA) reduce features into components that are linear combinations of the initial input space. However, the most intuitive inputs for the user are simple words like those contained in initial set of algorithmically selected features.

###Forward stepwise iterative regression

Before diving into identifying dimensions, I want to explore the space of features a bit more and get a sense of which features explain the most unique variance across climbs. to do this, I do an iterative regression analysis (take individual feature with highest R2 and then iterate with the residuals) on the climb x feature matrix

```python
%autoreload
avgs=ndf[descripfeats]
rx=mo.iterativeregression(avgs, plotit=True, limit=75)
```

Over 50% of the variance in the climb x attribute matrix is captured using the first 20 text features and >90 with the first 75 features. This suggests that there are redundant/collinear features and that further feature selection is warrented. We could just take the top N features from the iterative regression procedure, as this would provide a set of features that explain unique variance in the initial climb x attribute matrix. However, this approach does not tell us that these features are particularly informative of individual user prefernces. To get at informativeness more directly we can look at the feature importances from the random forest algorithm to identify features that are informative in classifying individual star ratings.

```python
#saving feature importances
%autoreload
clf=mo.classifier('rfc')
clf.n_estimators=nestimators
results=[]
resutlsdict={}
name='combined_feature_selection'
print "classifying using %s features" %(len(cfeatures)+len(sfeatures))
summarydf, resultsdf, informativefeats=mo.classify(clf,users, sfeatures, cfeatures, full_ndf, usesdf, minratings=minratings, dropself=True, getfeats=True)
resultsdict[name]=[summarydf, resultsdf]
results.append(summarydf['score'].mean())
```

```python
name='combined_feature_selection'
print "classifying with %s" %name
summarydf, resultsdf=resultsdict[name]
viz.plotresults(resultsdf, 'true_rating', 'feat_pred')
acc, precision, recall, f1=mo.round2score(resultsdf, 'true_rating', 'feat_pred', compare)
compare['name'].append(name)
```

####Feature selection approach:
    1. for each user, take top N most important features on each iteration, and comput proportion of folds in which the feature was informative
    2. compute one-sample T test across users to generate a tvalue reflecting how reliably informative a feature is
    3. take top 100 features as final feature space
    4. from this set of 100, used domain-knowledge to prune to 20 features to serve as a reduced feature space to use with new users

```python
featdf=informativefeats[[col for col in informativefeats.columns if col != 'climber']].T
featdf['T_value']=featdf.apply(lambda x:scipy.stats.ttest_1samp(x.values, 0)[0], axis=1)
featdf['P_value']=featdf.apply(lambda x:scipy.stats.ttest_1samp(x.values, 0)[1], axis=1)
sorteddf=featdf.sort(columns='T_value', ascending=False).iloc[:100,:]
```

###Feature pruning

We have 100 algorithmically selected features, but when we want to onboard new users, it won't make sense to ask them 100 questions to map them into this feature space. Also, not all the features make sense to ask as preference questions (i.e. "do you like great climbs" is an odd question). Thus, I reduce to smaller subset of 20 features that make sense to ask individual users and will use these as the reduced model space for new users. In contrast to the feature selection approaches used thus far, this final pruning from 100 to 20 features is performed using domain knowledge/product intuition. 

```python
askterms=['flakes', 'hand','tricky','undercling','slab','clipping','technical','boulder','bulge','bolted','steep', 'jug','chimney', 'crack','crux','roof','flake','cracks','traverse','pitch','crimp']
askfeats=[a+'_description' for a in askterms]
print askterms
```

###Assess predictive power of reduced text space alone
